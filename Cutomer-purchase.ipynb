EXPLORATORY DATA ANALYSIS

GENDER

```python
#plotting a bar chart for gender and it's count
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

Univariate:It shows that females are more then men

```python
# plotting a bar chart for gender vs total amount

sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gen,color='red')
'''for bars in sales_gen.containers:
    sales_gen.bar_label(bars)'''
```

Bivariate:This graph shows that females spend more amount than men

From above graphs we can see that most of the buyers are females and even the purchasing power of female are greater than men

Age

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

This graph shows that the age group 26-35 have the highest spending whereas 0-17 have the least spending

```python
#total amount vs age group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age,color='orange',palette='rocket')
```

This graph shows that the age group 26-35 has spent the most amount and 0-17 has done the least spending

```python
#total number of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
custom_color=sns.color_palette('husl',10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data=sales_state,x='State',y='Orders',palette=custom_color)
```

This graph shows that UP has placed the most number of orders and Harayana and Gujarat have the least orders placed
